```python
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv("mcdonalds_regression_sales_data.csv")
data.head(1)
```

```python
X = data.drop("Systemwide_Sales_Billions", axis=1)
y = data["Systemwide_Sales_Billions"]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.80, random_state = 42)
```

```python
import numpy as np
keras.backend.clear_session()
np.random.seed(42)
tf.random.set_seed(42)

model = Sequential()
model.add(Dense(10,activation="linear", input_dim=X_train.shape[1]))

model.add(Dense(10,activation="linear", input_dim=X_train.shape[1]))

model.add(Dense(10,activation="linear", input_dim=X_train.shape[1]))
model.add(Dense(10,activation="linear", input_dim=X_train.shape[1]))
model.add(Dense(10,activation="linear", input_dim=X_train.shape[1]))
model.add(Dense(10,activation="linear", input_dim=X_train.shape[1]))

model.add(Dense(10,activation="linear", input_dim=X_train.shape[1]))
model.add(Dense(10,activation="linear", input_dim=X_train.shape[1]))
model.add(Dense(10,activation="linear", input_dim=X_train.shape[1]))

model.add(Dense(10,activation="linear", input_dim=X_train.shape[1]))
model.add(Dense(1, activation="linear"))
optimizer = keras.optimizers.Adam(learning_rate=0.001)
a = model.compile(loss="mean_absolute_error", optimizer=optimizer)
a = fit = model.fit(X_train, y_train, epochs=100, verbose=1, batch_size=1)
```

```python
pred = model.predict(X_test)
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, model.predict(X_test))


```
